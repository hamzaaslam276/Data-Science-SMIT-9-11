```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generating a dataset with potential outliers
np.random.seed(0)
data = np.random.normal(100, 20, 200)
data = np.append(data, [300, 5])  # Adding potential outliers

# Calculating IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Defining the outlier criteria (1.5 times the IQR)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identifying outliers
outliers = (data < lower_bound) | (data > upper_bound)

# Removing outliers
data_no_outliers = data[~outliers]

# Plotting the original data with outliers
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=data)
plt.title('Original Data with Outliers')

# Plotting the data after removing outliers
plt.subplot(1, 2, 2)
sns.boxplot(data=data_no_outliers)
plt.title('Data After Removing Outliers')

plt.show()
```

```python
# Variance
import numpy as np
import matplotlib.pyplot as plt

# Example dataset
data = np.array([2, 4, 6, 8, 10])

# Calculate the mean and variance
mean = np.mean(data)
variance = np.var(data)

# Plotting
plt.figure(figsize=(8, 4))
plt.title("Mean and Variance of the Dataset")
plt.bar(['Mean', 'Variance'], [mean, variance], color=['blue', 'orange'])
plt.ylabel('Value')
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom, bernoulli, t, expon

# Setup figure
fig, axs = plt.subplots(3, 2, figsize=(14, 12))
fig.suptitle("Different Types of Distributions", fontsize=16)

# 1. Uniform Distribution
uniform_data = np.random.uniform(low=0, high=10, size=1000)
sns.histplot(uniform_data, bins=20, kde=True, color='g', ax=axs[0, 0], edgecolor='black')
axs[0, 0].set_title("Uniform Distribution")
axs[0, 0].set_xlabel("Value")
axs[0, 0].set_ylabel("Density")

# 2. Binomial Distribution
n, p = 10, 0.5
binomial_data = binom.rvs(n, p, size=1000)
sns.histplot(binomial_data, discrete=True, color='y', ax=axs[0, 1], edgecolor='black')
axs[0, 1].set_title("Binomial Distribution")
axs[0, 1].set_xlabel("Number of Successes")
axs[0, 1].set_ylabel("Frequency")

# 3. Bernoulli Distribution
p = 0.5
bernoulli_data = bernoulli.rvs(p, size=1000)
sns.histplot(bernoulli_data, discrete=True, color='skyblue', ax=axs[1, 0], edgecolor='black')
axs[1, 0].set_title("Bernoulli Distribution")
axs[1, 0].set_xlabel("Outcome (0 or 1)")
axs[1, 0].set_ylabel("Frequency")

# 4. Normal Distribution
mu, sigma = 0, 1
normal_data = np.random.normal(mu, sigma, 1000)
sns.histplot(normal_data, bins=20, kde=True, color='b', ax=axs[1, 1], edgecolor='black')
axs[1, 1].set_title("Normal Distribution")
axs[1, 1].set_xlabel("Value")
axs[1, 1].set_ylabel("Density")

# 5. T-Test Distribution
df = 10  # degrees of freedom
t_data = t.rvs(df, size=1000)
sns.histplot(t_data, bins=20, kde=True, color='r', ax=axs[2, 0], edgecolor='black')
axs[2, 0].set_title("T-Test Distribution")
axs[2, 0].set_xlabel("Value")
axs[2, 0].set_ylabel("Density")

# 6. Exponential Distribution
scale = 1  # scale = 1/lambda
exponential_data = expon.rvs(scale=scale, size=1000)
sns.histplot(exponential_data, bins=20, kde=True, color='k', ax=axs[2, 1], edgecolor='black')
axs[2, 1].set_title("Exponential Distribution")
axs[2, 1].set_xlabel("Value")
axs[2, 1].set_ylabel("Density")

# Adjust layout and show
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import beta, gamma

# Setup figure
fig, axs = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle("Gamma and Beta Distributions", fontsize=16)

# 1. Gamma Distribution
shape, scale = 2, 2  # alpha and beta
gamma_data = gamma.rvs(shape, scale=scale, size=1000)
sns.histplot(gamma_data, bins=20, kde=True, color='g', ax=axs[0], edgecolor='black')
axs[0].set_title("Gamma Distribution")
axs[0].set_xlabel("Value")
axs[0].set_ylabel("Density")

# 2. Beta Distribution
a, b = 2, 5  # alpha and beta
beta_data = beta.rvs(a, b, size=1000)
sns.histplot(beta_data, bins=20, kde=True, color='k', ax=axs[1], edgecolor='black')
axs[1].set_title("Beta Distribution")
axs[1].set_xlabel("Value")
axs[1].set_ylabel("Density")

# Adjust layout and show
plt.tight_layout(rect=[0, 0, 1, 0.94])
plt.show()
```

```python
df = sns.load_dataset('titanic')
df.head()
```

### Creat a contigency table

```python
contigency_table = pd.crosstab(df['sex'],df['survived'])
contigency_table
```

### perform chi-square test

```python
from scipy.stats import chi2_contingency
chi2,p,dof,expected = chi2_contingency(contigency_table)
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print(f"Expected:\n {expected}")

```

### as the p-value in this test is p-value: 1.1973570627755645e-58 which is far less than 0.05, we reject the Null hypothesis.

### shapiro wilk test on normal distribution gaussio Distribution Null Hypothesis: the data is normal distributed


```python
from scipy.stats import shapiro

# Perform Shapiro-Wilk test
stat, p = shapiro(df['age'])

# Print results
print("Test Statistic:", stat)
print("p-value:", p)

# Interpretation
if p > 0.05:
    print("Data is normally distributed (fail to reject null hypothesis).")
else:
    print("Data is not normally distributed (reject null hypothesis).")
```

### t_Test
1. One sample t-Test

```python
from scipy.stats import ttest_1samp

# Data
x = [1, 2, 3, 4, 5]
mu = 4

# Perform one-sample t-test
stat, p = ttest_1samp(x, mu)

# Print results
print("Test Statistic:", stat)
print("p-value:", p)

# Interpretation
if p > 0.05:
    print("Fail to reject the null hypothesis (mean is equal to mu).")
else:
    print("Reject the null hypothesis (mean is not equal to mu).")
```
