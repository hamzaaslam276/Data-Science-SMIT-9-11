```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x = [1,2,3,4,5]
y =[3,1,5,2,4]

plt.plot(x,y)

plt.xticks(x, labels=['Python','.Net','c++','angular','java'])
plt.yticks(x)

plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='y = sin(x)')
plt.title('Basic Axis Control')

plt.axis([0, 12, -1.5, 1.5])  # [xmin, xmax, ymin, ymax]

plt.legend()
plt.show()
```

```python
plt.plot(x, y, label='y = sin(x)')
plt.title('Axes Turned Off')

# Turn off the axes
plt.axis('off')

plt.legend()
plt.show()
```

```python
plt.plot(np.cos(x), np.sin(x))  
plt.title('Equal Aspect Ratio')

plt.axis('equal')

plt.show()
```

```python
x = np.linspace(1, 100, 100)
y = np.log(x)

plt.plot(x, y, label='y = log(x)')
plt.title('Logarithmic Axis')

plt.xscale('log')

plt.xlabel('Log-scaled X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.cos(x)

plt.plot(x, y, label='y = cos(x)')
plt.title('Custom Axis Ticks')

plt.xticks(ticks=[0, 2, 4, 6, 8, 10], labels=['0', '2π', '4π', '6π', '8π', '10π'])
plt.yticks(ticks=[-1, 0, 1], labels=['Min', 'Zero', 'Max'])

plt.legend()
plt.show()
```

```python
plt.plot(x, y, label='y = cos(x)')
plt.title('Axis Label Example')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

plt.legend()
plt.show()
```

```python
plt.plot(x, y, label='y = cos(x)')
plt.title('Set Axis Limits')

plt.xlim(2, 8) 
plt.ylim(-0.5, 1)

plt.legend()
plt.show()
```
