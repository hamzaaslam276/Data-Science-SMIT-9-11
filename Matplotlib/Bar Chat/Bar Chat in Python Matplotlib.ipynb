# Bar Chat in Python Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4]
y= [2,4,6,8]
c = ['r','g','k','y']
plt.bar(x,y, color = c)
plt.show()
```

```python
f = ['Python','.Net','Angular','SQL']
g = [90,85,80,75]
c= ('r', 'b', 'y','g')
plt.xlabel('Language', fontsize = 20)
plt.ylabel("No.", fontsize = 20)
plt.title("Language Graph", fontsize = 20)
plt.bar(f,g ,color=c,width=0.4,edgecolor='k',linewidth= 10, linestyle='-', alpha=.8,label="Popularity",)
plt.legend(fontsize='large',loc="upper right")
plt.show()
```

# Vertical bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np 

f = ['Python','.Net','Angular','SQL']
g = [90,85,80,75]
h = [30,40,50,60]
i = [10,15,20,25]
width = 0.2
p = np.arange(len(f))
p1 = [j+width for j in p]
p2 = [k+width for k in p1]
plt.xlabel('Language', fontsize = 20)
plt.ylabel("No.", fontsize = 20)
plt.title("Language Popularity Graph", fontsize = 20)
plt.bar(p,g,width,alpha=.8,label="Popularity")
plt.bar(p1,h,width,alpha=.8,label="Popularity1")
plt.bar(p2,i,width,alpha=.8,label="Popularity2")

# plt.xticks(p+width/2,f) # if use 2 bar than use this code 
plt.xticks(p+width,f,rotation=30)

plt.legend(fontsize='large',loc="upper center")
plt.show()
```

# Horizontal bar Chart 

```python
import matplotlib.pyplot as plt
import numpy as np 

f = ['Python','.Net','Angular','SQL']
g = [90,85,80,75]
h = [30,40,50,60]
i = [10,15,20,25]
height = 0.2
p = np.arange(len(f))
p1 = [j+height for j in p]
p2 = [k+width for k in p1]
plt.ylabel('Language', fontsize = 20)
plt.xlabel("No.", fontsize = 20)
plt.title("Language Popularity Graph", fontsize = 20)
plt.barh(p,g,height,alpha=.8,label="Popularity")
plt.barh(p1,h,height,alpha=.8,label="Popularity1")
plt.barh(p2,i,height,alpha=.8,label="Popularity2")

# plt.xticks(p+width/2,f) # if use 2 bar than use this code 
plt.yticks(p+width,f,rotation=30)

plt.legend(fontsize='large',loc="lower right")
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values1 = [3, 7, 2, 5]
values2 = [4, 6, 1, 4]

plt.bar(categories, values1, label='Group 1')
plt.bar(categories, values2, bottom=values1, label='Group 2')

plt.title('Stacked Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]
errors = [0.5, 0.2, 0.3, 0.4]

plt.bar(categories, values, yerr=errors, capsize=5, color='lightgreen')
plt.title('Bar Chart with Error Bars')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

total = sum(values)
percentages = [v / total * 100 for v in values]

plt.bar(categories, percentages, color='purple')"LinePlot in Python Matplotlib.ipynb"
plt.title('Bar Chart with Percentages')
plt.xlabel('Categories')
plt.ylabel('Percentage')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

sorted_categories = [x for _, x in sorted(zip(values, categories))]
sorted_values = sorted(values)

plt.bar(sorted_categories, sorted_values, color='cyan')
plt.title('Sorted Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values1 = [3, 7, 2, 5]
values2 = [4, 6, 1, 4]

plt.bar(categories, values1, alpha=0.5, label='Values 1')
plt.bar(categories, values2, alpha=0.5, label='Values 2')

plt.title('Multiple Bar Charts on Same Axes')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

plt.bar(categories, values)

plt.xticks(categories, ['Category A', 'Category B', 'Category C', 'Category D'])
plt.title('Bar Chart with Custom Axis Ticks')
plt.xlabel('Custom Categories')
plt.ylabel('Values')
plt.show()
```
