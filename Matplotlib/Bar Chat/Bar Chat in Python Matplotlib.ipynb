```python
f = ['Python','.Net','Angular','SQL']
g = [90,85,80,75]
c= ('r', 'b', 'y','g')
plt.xlabel('Language', fontsize = 20)
plt.ylabel("No.", fontsize = 20)
plt.title("Language Graph", fontsize = 20)
plt.bar(f,g ,color=c,width=0.4,edgecolor='k',linewidth= 10, linestyle='-', alpha=.8,label="Popularity",)
plt.legend(fontsize='large',loc="upper right")
plt.show()
```

# Vertical bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np 

f = ['Python','.Net','Angular','SQL']
g = [90,85,80,75]
h = [30,40,50,60]
i = [10,15,20,25]
width = 0.2
p = np.arange(len(f))
p1 = [j+width for j in p]
p2 = [k+width for k in p1]
plt.xlabel('Language', fontsize = 20)
plt.ylabel("No.", fontsize = 20)
plt.title("Language Popularity Graph", fontsize = 20)
plt.bar(p,g,width,alpha=.8,label="Popularity")
plt.bar(p1,h,width,alpha=.8,label="Popularity1")
plt.bar(p2,i,width,alpha=.8,label="Popularity2")

# plt.xticks(p+width/2,f) # if use 2 bar than use this code 
plt.xticks(p+width,f,rotation=30)

plt.legend(fontsize='large',loc="upper center")
plt.show()
```

# Horizontal bar Chart 

```python
import matplotlib.pyplot as plt
import numpy as np 

f = ['Python','.Net','Angular','SQL']
g = [90,85,80,75]
h = [30,40,50,60]
i = [10,15,20,25]
height = 0.2
p = np.arange(len(f))
p1 = [j+height for j in p]
p2 = [k+width for k in p1]
plt.ylabel('Language', fontsize = 20)
plt.xlabel("No.", fontsize = 20)
plt.title("Language Popularity Graph", fontsize = 20)
plt.barh(p,g,height,alpha=.8,label="Popularity")
plt.barh(p1,h,height,alpha=.8,label="Popularity1")
plt.barh(p2,i,height,alpha=.8,label="Popularity2")

# plt.xticks(p+width/2,f) # if use 2 bar than use this code 
plt.yticks(p+width,f,rotation=30)

plt.legend(fontsize='large',loc="lower right")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)  
categories = ['A', 'B', 'C', 'D'] 
values = [5, 7, 3, 8]  
data = np.random.randn(1000)

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].plot(x, y, color='b', label='Sine Wave')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].legend()

axs[0, 1].bar(categories, values, color='g', alpha=0.7)
axs[0, 1].set_title('Bar Plot')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

scatter_sizes = np.random.rand(100) * 100 
axs[1, 0].scatter(x, y, c='r', s=scatter_sizes, alpha=0.5, edgecolor='k', label='Scatter Points')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')
axs[1, 0].legend()

axs[1, 1].hist(data, bins=20, color='m', alpha=0.7, edgecolor='k')
axs[1, 1].set_title('Histogram')
axs[1, 1].set_xlabel('Bins')
axs[1, 1].set_ylabel('Frequency')

plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)  
categories = ['A', 'B', 'C', 'D'] 
values = [5, 7, 3, 8]  
data = np.random.randn(1000)


plt.subplot(2,2,1)
plt.plot(x, y, color='b', label='Sine Wave')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()


plt.subplot(2,2,2)
plt.bar(categories, values, color='g', alpha=0.7)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.subplot(2,2,3)
scatter_sizes = np.random.rand(100) * 100 
plt.scatter(x, y, c='r', s=scatter_sizes, alpha=0.5, edgecolor='k', label='Scatter Points')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.subplot(2,2,4)
plt.hist(data, bins=20, color='m', alpha=0.7, edgecolor='k')
plt.title('Histogram')
plt.xlabel('Bins')
plt.ylabel('Frequency')

plt.tight_layout()

plt.show()


```
