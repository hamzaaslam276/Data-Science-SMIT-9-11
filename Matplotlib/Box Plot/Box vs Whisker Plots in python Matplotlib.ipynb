```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x = [10,20,30,40,50,120,200]
plt.boxplot(x,vert=True,widths=0.2,labels=['Python'],patch_artist=True)

plt.show()
```

```python
x = [10,20,30,40,50,120,200]
plt.boxplot(x,vert=True,widths=0.2,labels=['Python'],patch_artist=True,sym='g+')

plt.show()
```

```python
x = [10,20,30,40,50,120,200]
plt.boxplot(x,notch=False,widths=0.2,labels=['Python'],showmeans=True,whis=2)

plt.show()
```

```python
x = [10,20,30,40,50]
plt.boxplot(x,notch=True,widths=0.2,labels=['Python'],showmeans=True)

plt.show()
```

```python
x = [10,20,30,40,50]
plt.boxplot(x,labels=['Python'],showmeans=True,sym='g+',boxprops=dict(color='r'),capprops=dict(color='b'),whiskerprops=dict(color='m'))

plt.show()
```

```python
x = [10,20,30,40,50]
x1 = [20,25,35,45,55]
y = [x,x1]
plt.boxplot(y,labels=['Python','.Net'],showmeans=True,sym='g+',boxprops=dict(color='r'),capprops=dict(color='b'),whiskerprops=dict(color='m'))

plt.show()
```

```python
x = [10,20,30,40,50]
x1 = [20,25,35,45,55]
y = [x,x1]
plt.boxplot(y,vert=False,labels=['Python','.Net'],showmeans=True,sym='g+',boxprops=dict(color='r'),capprops=dict(color='b'),whiskerprops=dict(color='m'))

plt.show()
```

```python
data = np.random.rand(100, 1)

plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor='lightgreen', color='blue'), whiskerprops=dict(color='green'), capprops=dict(color='red'))
plt.title('Box Plot with Custom Colors')
plt.ylabel('Values')
plt.show()
```

```python
data = np.random.rand(100, 1)
data = np.append(data, [5, 6, 7])  

plt.boxplot(data)
plt.title('Box Plot with Outliers')
plt.ylabel('Values')
plt.show()
```

```python
data = np.random.rand(100, 1)

plt.boxplot(data, 
            vert=True, 
            patch_artist=True, 
            boxprops=dict(facecolor='lightblue', color='black'), 
            whiskerprops=dict(color='purple', linewidth=1.5), 
            capprops=dict(color='orange', linewidth=2), 
            flierprops=dict(marker='o', color='red', markersize=5), 
            medianprops=dict(color='green', linewidth=2))
plt.title('Customized Box Plot')
plt.ylabel('Values')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data1 = np.random.rand(100, 1).flatten()  
data2 = np.random.rand(100, 1).flatten() + 10
data3 = np.random.rand(100, 1).flatten() + 100

plt.boxplot([data1, data2, data3])
plt.title('Box Plot for Data with Different Scales')
plt.ylabel('Values')
plt.show()
```

```python
data = np.random.lognormal(mean=0, sigma=1, size=1000)

plt.boxplot(data)
plt.yscale('log')
plt.title('Box Plot with Logarithmic Scale')
plt.ylabel('Values')
plt.show()
```
