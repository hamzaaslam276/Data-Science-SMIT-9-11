```python
import matplotlib.pyplot as plt 
```

```python
x= [1,2,3,4,5]
y= [2,5,3,7,4]

plt.step(x,y,marker='o',color='r',ms=10,mfc='g',label='Python')
plt.title("Python",fontsize = 16)
plt.xlabel("X-axis",fontsize = 14)
plt.ylabel("Y-axis",fontsize = 14)
plt.legend(loc=2)
plt.fill_between(x,y)

plt.show()
```

```python
x= [1,2,3,4,5]
y= [1,2,3,4,5]

plt.title("Python",fontsize = 16)
plt.xlabel("X-axis",fontsize = 14)
plt.ylabel("Y-axis",fontsize = 14)
plt.plot(x,y)

plt.fill_between(x,y)
plt.show()
```

```python
x= [1,2,3,4,5]
y= [1,2,3,4,5]

plt.title("Python",fontsize = 16)
plt.xlabel("X-axis",fontsize = 14)
plt.ylabel("Y-axis",fontsize = 14)
plt.plot(x,y,color = 'r')

plt.fill_between(x=[2,4], y1=2, y2=4,color='g')
plt.show()
```

```python
import numpy as np 
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.title("Python",fontsize = 16)
plt.xlabel("X-axis",fontsize = 14)
plt.ylabel("Y-axis",fontsize = 14)
plt.plot(x,y,color = 'r')

plt.fill_between(x,y,color='g',where=(x>2) & (x<4))
plt.show()
```

```python
x = np.linspace(0, 10, 100)  
y = np.sin(x)

plt.plot(x, y, label='y = sin(x)', color='b')

plt.fill_between(x, y, color='lightblue', alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Filling Between Curve and X-axis')
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='y = sin(x)', color='k')
plt.plot(x, y2, label='y = cos(x)', color='r')

plt.fill_between(x, y1, y2, color='purple', alpha=0.3, label='Area Between Curves')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Filling Between Two Curves')
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='y = sin(x)', color='blue')

plt.fill_between(x, y, where=(y >= 0), color='g', alpha=0.5, label='y >= 0')
plt.fill_between(x, y, where=(y < 0), color='r', alpha=0.5, label='y < 0')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Conditional Filling')
plt.legend()
plt.show()
```
