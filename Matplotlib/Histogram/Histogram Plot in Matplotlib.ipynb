```python
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2  

plt.hist(data1, bins=30, alpha=0.5, label='Data 1')
plt.hist(data2, bins=30, alpha=0.5, label='Data 2')
plt.title('Multiple Histograms')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram with Custom Color')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, cumulative=True, color='purple', edgecolor='black')
plt.title('Cumulative Histogram')
plt.xlabel('Value')
plt.ylabel('Cumulative Frequency')
plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, density=True, color='green', edgecolor='black')
plt.title('Normalized Histogram (Density)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2 

plt.hist([data1, data2], bins=30, stacked=True, color=['blue', 'red'], edgecolor='black')
plt.title('Stacked Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
data = np.random.chisquare(2, 1000)

plt.hist(data, bins=30, color='brown', edgecolor='black')
plt.yscale('log')
plt.title('Histogram with Logarithmic Scale')
plt.xlabel('Value')
plt.ylabel('Log(Frequency)')
plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, range=(-3, 3), edgecolor='black')
plt.title('Histogram with Custom Range')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
data1 = np.random.randn(1000)           
data2 = np.random.uniform(-3, 3, 1000)   

plt.hist(data1, bins=30, alpha=0.5, label='Normal Distribution', edgecolor='black')
plt.hist(data2, bins=30, alpha=0.5, label='Uniform Distribution', edgecolor='black')
plt.title('Histograms of Different Data Distributions')
plt.annotationsxlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2 

plt.hist(data1, bins=10, alpha=0.5, label='Data 1 (10 bins)', edgecolor='black')
plt.hist(data2, bins=50, alpha=0.5, label='Data 2 (50 bins)', edgecolor='black')
plt.title('Histograms with Different Binning Strategies')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
