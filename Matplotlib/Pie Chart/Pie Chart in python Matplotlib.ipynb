```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x1 = [10,20,30,40]
x2 = ["A","B","C","D"]

plt.pie(x1,x2)

plt.show()
```

```python
x = [30,10,40,20]
y = ['.Net','angular', 'Python', 'java']
ex = [0.0,0.2,0.0,0.0]
plt.title('Pie chart in python Matplotlib',fontsize= 20)
plt.pie(x,labels=y,explode=ex,autopct='%0.1f%%',shadow=True,radius=1,textprops={'fontsize':15},rotatelabels=True,labeldistance=0.7)
plt.show()
```

```python
x = [30,10,40,20]
y = ['.Net','angular', 'Python', 'java']
ex = [0.0,0.3,0.0,0.0]

plt.title('Pie chart in python Matplotlib',fontsize= 20)
plt.pie(x,labels=y,rotatelabels=False,explode=ex,autopct='%0.1f%%',textprops={'fontsize':15},wedgeprops={'linewidth':4,'edgecolor':'m'},)
plt.legend(loc="center left",)

plt.show()
```

```python
plt.pie([1])

plt.title('Dot chart in python Matplotlib',fontsize= 20)
plt.show()
```

```python
x = [30,10,40,20]
x1 = [20,40,10,30]
y = ['.Net','angular', 'Python', 'java']
y1 = ['c','SQL','c#','c++']


plt.title('Multi Pie chart in python Matplotlib',fontsize= 20)
plt.pie(x,labels=y,radius=1,autopct="%.1f%%",labeldistance=0.7,)
plt.pie(x1,labels= y1,radius=.5,autopct="%.1f%%",labeldistance=0.3)
plt.legend(loc=1,)

plt.show()
```

```python
x = [30,10,40,20]
y = ['.Net','angular', 'Python', 'java']


plt.title('Ring Pie chart in python Matplotlib',fontsize= 20)
plt.pie(x,labels=y,radius=1,labeldistance=.7,)
plt.pie([1],radius=.7,colors="w")
plt.legend(loc=1,)

plt.show()
```

```python
x = [30,10,40,20]
y = ['.Net','angular', 'Python', 'java']
c = plt.Circle(xy=(0,0),radius=.5,facecolor='w')

plt.title('Ring Pie chart in python Matplotlib',fontsize= 20)
plt.pie(x,labels=y,radius=1,labeldistance=.7,)
plt.gca().add_artist(c)


plt.legend(loc=1,)
plt.show()
```

```python
labels = ['.Net','angular', 'Python', 'java']
sizes = [15, 30, 40, 15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Programming Languages Popularity (Equal Aspect Ratio)')
plt.axis('equal')  
plt.show()
```
