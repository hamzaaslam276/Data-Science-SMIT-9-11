```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x  = [1,2,3,4,5]
y = [2,3,1,4,5]

plt.plot(x,y,color= 'r')

plt.savefig("Line",dpi = 2000)

plt.show()
```

```python
x  = [1,2,3,4,5]
y = [2,3,1,4,5]

plt.plot(x,y,color= 'r')

plt.savefig("Line.pdf",dpi = 2000,facecolor='g')

plt.show()
```

```python
x  = [1,2,3,4,5]
y = [2,3,1,4,5]

plt.plot(x,y,color= 'r')

plt.savefig("Line.pdf",dpi = 2000,facecolor='g',transparent=True)

plt.show()
```

```python
x  = [1,2,3,4,5]
y = [2,3,1,4,5]

plt.plot(x,y,color= 'r')

plt.savefig("Line",dpi = 2000,facecolor='g',transparent=True,bbox_inches='tight')

plt.show()
```

```python
x = np.linspace(0, 10, 100)
for i in range(3):
    y = np.sin(x + i)
    plt.plot(x, y, label=f'Shift {i}')
    plt.title(f'Sine Wave Shifted by {i}')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    
    plt.savefig(f'sine_wave_shift_{i}.jpg', dpi=150)
    plt.clf()  # Clear the current figure to avoid overlap
```
