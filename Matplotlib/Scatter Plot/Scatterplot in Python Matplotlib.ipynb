```python
import matplotlib.pyplot as plt 

day = [1,2,3,4,5,6,7]
no = [3,5,7,4,2,1,6]

plt.title("Scatter Plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No.",fontsize = 15)
sizes = [200,300,400,500,100,190,250]
colors = [20,49,55,62,77,81,39]
plt.scatter(day,no,edgecolors="r",marker="*",linewidths=1,s=sizes,c=colors,cmap="viridis") # color(Accent,BrBG)
t= plt.colorbar()
t.set_label("Colorbar",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt 

day = [1,2,3,4,5,6,7]
no = [3,5,7,4,2,1,6]
no1 = [7,4,2,1,6,5,3]

plt.title("Scatter Plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No.",fontsize = 15)
sizes = [200,300,400,500,100,190,250]
colors = [20,49,55,62,77,81,39]
plt.scatter(day,no,edgecolors="r",marker="*",linewidths=1,s=sizes,c=colors,cmap="viridis") # color(Accent,BrBG)
plt.scatter(day,no1,edgecolors="r",s=sizes,c=colors,cmap="viridis") # color(Accent,BrBG)
t= plt.colorbar()
t.set_label("Colorbar",fontsize=15)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt



x = np.random.rand(100) * 100  
y = np.random.rand(100) * 100  
sizes = np.random.rand(100) * 300 
conditions = x + y > 100
colors = np.where(conditions, 'r', 'k')

plt.scatter(x, y, s=sizes,c=colors ,edgecolor='k',marker="*")

plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.title('Scatter Plot of Random Data Points', fontsize=16)


plt.scatter([], [], c='r', label='x + y > 100',marker="*")
plt.scatter([], [], c='k', label='x + y <= 100',marker="*")
plt.legend(title="Point Categories", fontsize=12,loc=1)

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
x = np.random.rand(100, 1)
y = 2 * x + 1 + np.random.randn(100, 1) * 0.2  

plt.scatter(x, y, color='blue', alpha=0.5, label='Data')

model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)

plt.plot(x, y_pred, color='red', label='Regression Line')

plt.title('Scatter Plot with Regression Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
x = np.logspace(0.1, 2, 50)
y = np.logspace(0.1, 2, 50)

plt.scatter(x, y)
plt.xscale('log')
plt.yscale('log')
plt.title('Scatter Plot with Logarithmic Scales')
plt.xlabel('X-axis (Log Scale)')
plt.ylabel('Y-axis (Log Scale)')
plt.show()
```

```python
x1 = np.random.rand(50)
y1 = np.random.rand(50)
x2 = np.random.rand(50) + 1
y2 = np.random.rand(50) + 1

plt.scatter(x1, y1, color='red', label='Group 1', alpha=0.5)
plt.scatter(x2, y2, color='blue', label='Group 2', alpha=0.5)

plt.title('Scatter Plot with Multiple Groups')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
x = np.random.rand(10)
y = np.random.rand(10)

plt.scatter(x, y)

for i in range(len(x)):
    plt.text(x[i], y[i], f'({x[i]:.2f}, {y[i]:.2f})', fontsize=9, ha='right')

plt.title('Scatter Plot with Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='green', marker='o')

ax.set_title('3D Scatter Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```
