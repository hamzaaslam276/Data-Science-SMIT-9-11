```python
import matplotlib.pyplot as plt 

day = [1,2,3,4,5,6,7]
no = [3,5,7,4,2,1,6]

plt.title("Scatter Plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No.",fontsize = 15)
sizes = [200,300,400,500,100,190,250]
colors = [20,49,55,62,77,81,39]
plt.scatter(day,no,edgecolors="r",marker="*",linewidths=1,s=sizes,c=colors,cmap="viridis") # color(Accent,BrBG)
t= plt.colorbar()
t.set_label("Colorbar",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt 

day = [1,2,3,4,5,6,7]
no = [3,5,7,4,2,1,6]
no1 = [7,4,2,1,6,5,3]

plt.title("Scatter Plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No.",fontsize = 15)
sizes = [200,300,400,500,100,190,250]
colors = [20,49,55,62,77,81,39]
plt.scatter(day,no,edgecolors="r",marker="*",linewidths=1,s=sizes,c=colors,cmap="viridis") # color(Accent,BrBG)
plt.scatter(day,no1,edgecolors="r",s=sizes,c=colors,cmap="viridis") # color(Accent,BrBG)
t= plt.colorbar()
t.set_label("Colorbar",fontsize=15)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt



x = np.random.rand(100) * 100  
y = np.random.rand(100) * 100  
sizes = np.random.rand(100) * 300 
conditions = x + y > 100
colors = np.where(conditions, 'r', 'k')

plt.scatter(x, y, s=sizes,c=colors ,edgecolor='k',marker="*")

plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.title('Scatter Plot of Random Data Points', fontsize=16)


plt.scatter([], [], c='r', label='x + y > 100',marker="*")
plt.scatter([], [], c='k', label='x + y <= 100',marker="*")
plt.legend(title="Point Categories", fontsize=12,loc=1)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)  
categories = ['A', 'B', 'C', 'D'] 
values = [5, 7, 3, 8]  
data = np.random.randn(1000)

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 1].plot(x, y, color='b', label='Sine Wave')
axs[0, 1].set_title('Line Plot')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')
axs[0, 1].legend()

axs[0, 0].bar(categories, values, color='g', alpha=0.7)
axs[0, 0].set_title('Bar Plot')
axs[0, 0].set_xlabel('Categories')
axs[0, 0].set_ylabel('Values')

scatter_sizes = np.random.rand(100) * 100 
axs[1, 0].scatter(x, y, c='r', s=scatter_sizes, alpha=0.5, edgecolor='k', label='Scatter Points')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')
axs[1, 0].legend()

axs[1, 1].hist(data, bins=20, color='m', alpha=0.7, edgecolor='k')
axs[1, 1].set_title('Histogram')
axs[1, 1].set_xlabel('Bins')
axs[1, 1].set_ylabel('Frequency')

# plt.tight_layout()

plt.show()
```
