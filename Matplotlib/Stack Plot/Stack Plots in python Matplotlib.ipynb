```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x= [1,2,3,4,5]
y = [6,7,8,9,10]
plt.stackplot(x,y)
plt.title("Python",fontsize = 16)
plt.xlabel("X-axis",fontsize = 14)
plt.ylabel("Y-axis",fontsize = 14)
plt.grid()
plt.show()
```

```python
x= [1,2,3,4,5]
area1 = [1,7,5,2,9]
area2 = [2,6,4,1,7]
area3 = [1,5,9,3,7]

l = ['area1', 'area2', 'area3']
plt.stackplot(x,area1,area2,area3,labels=l)
plt.title("Python",fontsize = 16)
plt.xlabel("X-axis",fontsize = 14)
plt.ylabel("Y-axis",fontsize = 14)
plt.legend(loc=2)
plt.show()
```

```python
x= [1,2,3,4,5]
area1 = [1,7,5,7,9]
area2 = [2,6,4,10,7]
area3 = [1,5,9,4,7]

l = ['area1', 'area2', 'area3']
plt.stackplot(x,area1,area2,area3,labels=l,colors=['y','r','k'])
plt.title("Python",fontsize = 16)
plt.xlabel("X-axis",fontsize = 14)
plt.ylabel("Y-axis",fontsize = 14)
plt.legend(loc=2)
plt.show()
```

```python
x= [1,2,3,4,5]
area1 = [1,7,5,7,9]
area2 = [2,6,4,10,7]
area3 = [1,5,9,4,7]

l = ['area1', 'area2', 'area3']
plt.stackplot(x,area1,area2,area3,labels=l,baseline='sym')
plt.title("Python",fontsize = 16)
plt.xlabel("X-axis",fontsize = 14)
plt.ylabel("Y-axis",fontsize = 14)
plt.legend(loc=2)
plt.show()
```

```python
x= [1,2,3,4,5]
area1 = [1,7,5,7,9]
area2 = [2,6,4,10,7]
area3 = [1,5,9,4,7]

l = ['area1', 'area2', 'area3']
plt.stackplot(x,area1,area2,area3,labels=l,colors=['g','m','y'],baseline='wiggle')
plt.title("Python",fontsize = 16)
plt.xlabel("X-axis",fontsize = 14)
plt.ylabel("Y-axis",fontsize = 14)
plt.legend(loc=2)
plt.show()
```

```python
x = np.arange(1, 6)
y1 = np.array([3, 5, 1, 2, 3])
y2 = np.array([2, 3, 4, 3, 4])
y3 = np.array([1, 2, 3, 1, 2])

plt.stackplot(x, y1, y2, y3, labels=['A', 'B', 'C'], colors=['skyblue', 'orange', 'green'])
plt.title('Basic Stacked Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='upper left')
plt.show()
```

```python
x = np.arange(1, 6)
y1 = np.array([3, 4, 1, 2, 3])
y2 = np.array([1, 2, 3, 4, 5])
y3 = np.array([2, 3, 1, 3, 2])

plt.stackplot(x, y1, y2, y3, labels=['X', 'Y', 'Z'], alpha=0.7, colors=['purple', 'cyan', 'pink'])
plt.title('Stacked Plot with Transparency')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='upper left')
plt.show()
```

```python
x = np.arange(1, 6)
y1 = np.array([3, 5, 2, -2, 1])
y2 = np.array([1, 2, 3, -1, -3])
y3 = np.array([2, -1, 1, 3, 2])

plt.stackplot(x, y1, y2, y3, labels=['P', 'Q', 'R'], colors=['red', 'blue', 'green'])
plt.title('Stacked Plot with Negative Values')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='lower right')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  
plt.show()
```

```python
x = np.arange(1, 6)
y1 = np.array([1, 2, 3, 4, 5])
y2 = np.array([2, 3, 4, 5, 6])
y3 = np.array([3, 4, 5, 6, 7])

plt.stackplot(x, y1, y2, y3, labels=['Alpha', 'Beta', 'Gamma'], colors=['red', 'yellow', 'blue'])
plt.title('Stacked Plot with Grid')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='upper left')
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(8, 8))

x = np.arange(1, 6)
y1 = np.array([2, 4, 6, 8, 10])
y2 = np.array([1, 3, 5, 7, 9])
y3 = np.array([3, 6, 9, 12, 15])

axs[0].stackplot(x, y1, y2, y3, labels=['One', 'Two', 'Three'], colors=['lightblue', 'orange', 'green'])
axs[0].set_title('Stacked Plot 1')
axs[0].legend(loc='upper left')

axs[1].stackplot(x, y3, y2, y1, labels=['Three', 'Two', 'One'], colors=['pink', 'purple', 'yellow'])
axs[1].set_title('Stacked Plot 2')
axs[1].legend(loc='upper left')

plt.tight_layout()
plt.show()
```

```python
x = np.arange(1, 6)
y1 = np.array([3, 5, 2, 1, 4])
y2 = np.array([1, 2, 1, 2, 3])
y3 = np.array([2, 1, 3, 4, 2])

total = y1 + y2 + y3
y1_norm = y1 / total
y2_norm = y2 / total
y3_norm = y3 / total

plt.stackplot(x, y1_norm, y2_norm, y3_norm, labels=['A', 'B', 'C'], colors=['green', 'blue', 'red'])
plt.title('Stacked Plot with Normalized Data')
plt.xlabel('X-axis')
plt.ylabel('Proportion')
plt.legend(loc='upper left')
plt.show()
```
