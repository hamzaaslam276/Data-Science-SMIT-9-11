```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [5,3,2,1,4]

plt.stem(x,y)

plt.show()
```

```python
x = [1,2,3,4,5]
y = [5,3,2,1,4]

plt.stem(x,y,linefmt='k-',markerfmt="g*",bottom=2.5)

plt.show()
```

```python
x = [1,2,3,4,5]
y = [5,3,2,1,4]

plt.stem(x,y,linefmt='-',markerfmt="r+",bottom=0,basefmt="y",label="Python")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [5, 3, 2, 1, 4]

plt.stem(x, y, linefmt='m-', markerfmt="r", bottom=0, basefmt="k", label="Python", orientation="horizontal")
plt.legend()
plt.show()
```

```python
x = np.linspace(0.1, 2 * np.pi, 10)
y = np.sin(x)

plt.stem(x, y)
plt.title('Simple Stem Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = np.linspace(0.1, 2 * np.pi, 10)
y = np.sin(x)

plt.stem(x, y, linefmt='--', markerfmt='o', basefmt='-')
plt.title('Custom Line and Marker Styles')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = np.arange(1, 6)
y = np.array([2, 3, 5, 7, 11])

plt.stem(y, x, orientation='horizontal', linefmt='g-', markerfmt='ro', basefmt='k-')
plt.title('Horizontal Stem Plot')
plt.xlabel('Y-axis')
plt.ylabel('X-axis')
plt.show()
```

```python
x = np.arange(1, 6)
y = np.array([2, 3, 5, 7, 11])

plt.stem(x, y, linefmt='b-', markerfmt='go', basefmt='r-')
plt.title('Stem Plot with Labels')

for i, (xi, yi) in enumerate(zip(x, y)):
    plt.text(xi, yi + 0.5, f'({xi},{yi})', ha='center', fontsize=10)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = np.linspace(0.1, 2 * np.pi, 10)
y = np.sin(x)

plt.stem(x, y, linefmt='m-', markerfmt='ro', basefmt='k-')
plt.title('Stem Plot with Negative Values')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(6, 8))


x = np.linspace(0.1, 2 * np.pi, 10)
y1 = np.sin(x)
y2 = np.cos(x)

axs[0].stem(x, y1, linefmt='b-', markerfmt='go', basefmt='k-')
axs[0].set_title('Sine Stem Plot')

axs[1].stem(x, y2, linefmt='r-', markerfmt='bo', basefmt='k-')
axs[1].set_title('Cosine Stem Plot')

plt.tight_layout()
plt.show()
```

```python
x = np.arange(1, 6)
y = np.array([1, -2, 3, -4, 5])

plt.stem(x, y, linefmt='g-', markerfmt='bo', basefmt='k-')
plt.title('Highlight Specific Stems')

plt.stem([3], [3], linefmt='r-', markerfmt='ro')  

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
