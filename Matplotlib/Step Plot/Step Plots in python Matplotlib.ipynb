```python
import matplotlib.pyplot as plt
```

```python
x= [1,2,3,4,5]
y= [1,2,3,4,5]

plt.title("Python",fontsize = 16)
plt.xlabel("X-axis",fontsize = 14)
plt.ylabel("Y-axis",fontsize = 14)
plt.step(x,y)
plt.grid()
plt.show()
```

```python
x= [1,2,3,4,5]
y= [2,5,3,7,4]

plt.step(x,y,marker='o',color='r',ms=10,mfc='g',label='Python')
plt.title("Python",fontsize = 16)
plt.xlabel("X-axis",fontsize = 14)
plt.ylabel("Y-axis",fontsize = 14)
plt.legend(loc=2)
plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 0, 1, 0, 1]

plt.step(x, y, where='mid', color='b', linestyle='--', label='Python')

plt.title("Python",fontsize = 16)
plt.xlabel("X-axis",fontsize = 14)
plt.ylabel("Y-axis",fontsize = 14)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 1)  
y = np.cumsum(np.random.randint(1, 10, size=10)) 


''' Line 2: y = np.cumsum(np.random.randint(1, 10, size=10))
Purpose: Generates a cumulative sum of random integers between 1 and 9, creating an array of increasing values.

Steps:

np.random.randint(1, 10, size=10):
Generates an array of 10 random integers between 1 (inclusive) and 10 (exclusive).
Example output:
[3, 7, 2, 8, 1, 6, 5, 4, 9, 2]
np.cumsum(...):
Computes the cumulative sum of the array.
Example cumulative sum for the above random integers:
[3, 10, 12, 20, 21, 27, 32, 36, 45, 47]
'''


plt.step(x, y, where='pre', color='g', linestyle='-', linewidth=2, label='Python')

plt.grid(True, linestyle='--', alpha=0.6)
plt.title("Python",fontsize = 16)
plt.xlabel("X-axis",fontsize = 14)
plt.ylabel("Y-axis",fontsize = 14)
plt.legend()

plt.show()
```
