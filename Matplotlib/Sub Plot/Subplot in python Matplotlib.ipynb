```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(1, 2, 1)  
plt.plot(x, y1, color='blue', label='y = sin(x)')
plt.title('Sine Curve')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, y2, color='red', label='y = cos(x)')
plt.title('Cosine Curve')
plt.legend()

plt.tight_layout()  
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

plt.subplot(2, 2, 1)
plt.plot(x, y1, color='b')
plt.title('Sine')

plt.subplot(2, 2, 2) 
plt.plot(x, y2, color='g')
plt.title('Cosine')

plt.subplot(2, 2, 3) 
plt.plot(x, y3, color='r')
plt.title('Tangent')
plt.ylim(-10, 10) 

plt.subplot(2, 2, 4) 
plt.plot(x, y4, color='m')
plt.title('Exponential Decay')

plt.tight_layout() 
plt.show()
```

```python
fig, axes = plt.subplots(2, 1, sharex=True, figsize=(8, 6))

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

axes[0].plot(x, y1, color='blue', label='Sine')
axes[0].set_title('Sine Curve')
axes[0].legend()

axes[1].plot(x, y2, color='red', label='Cosine')
axes[1].set_title('Cosine Curve')
axes[1].legend()

fig.text(0.5, 0.04, 'X-axis', ha='center', fontsize=12)
fig.text(0.04, 0.5, 'Y-axis', va='center', rotation='vertical', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
from matplotlib.gridspec import GridSpec

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

fig = plt.figure(figsize=(8, 6))
gs = GridSpec(2, 2, figure=fig)

ax1 = fig.add_subplot(gs[0, 0])
ax1.plot(x, y1, color='blue')
ax1.set_title('Sine')

ax2 = fig.add_subplot(gs[0, 1]) 
ax2.plot(x, y2, color='green')
ax2.set_title('Cosine')

ax3 = fig.add_subplot(gs[1, :])  
ax3.plot(x, y3, color='orange')
ax3.set_title('Tangent')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)  
categories = ['A', 'B', 'C', 'D'] 
values = [5, 7, 3, 8]  
data = np.random.randn(1000)

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 1].plot(x, y, color='b', label='Sine Wave')
axs[0, 1].set_title('Line Plot')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')
axs[0, 1].legend()

axs[0, 0].bar(categories, values, color='g', alpha=0.7)
axs[0, 0].set_title('Bar Plot')
axs[0, 0].set_xlabel('Categories')
axs[0, 0].set_ylabel('Values')

scatter_sizes = np.random.rand(100) * 100 
axs[1, 0].scatter(x, y, c='r', s=scatter_sizes, alpha=0.5, edgecolor='k', label='Scatter Points')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')
axs[1, 0].legend()

axs[1, 1].hist(data, bins=20, color='m', alpha=0.7, edgecolor='k')
axs[1, 1].set_title('Histogram')
axs[1, 1].set_xlabel('Bins')
axs[1, 1].set_ylabel('Frequency')

# plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)  
categories = ['A', 'B', 'C', 'D'] 
values = [5, 7, 3, 8]  
data = np.random.randn(1000)


plt.subplot(2,2,1)
plt.plot(x, y, color='b', label='Sine Wave')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()


plt.subplot(2,2,2)
plt.bar(categories, values, color='g', alpha=0.7)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.subplot(2,2,3)
scatter_sizes = np.random.rand(100) * 100 
plt.scatter(x, y, c='r', s=scatter_sizes, alpha=0.5, edgecolor='k', label='Scatter Points')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.subplot(2,2,4)
plt.hist(data, bins=20, color='m', alpha=0.7, edgecolor='k')
plt.title('Histogram')
plt.xlabel('Bins')
plt.ylabel('Frequency')

plt.tight_layout()

plt.show()


```
