```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='y = sin(x)')
plt.title('Adding Text to Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.text(5, 0, 'This is the origin', fontsize=12, color='red')

plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='y = sin(x)')
plt.title('Annotating a Point')

plt.annotate(
    'Max Point',
    xy=(np.pi / 2, 1),  
    xytext=(2, 1.5),   
    arrowprops=dict(facecolor='blue', shrink=0.05),
    fontsize=12,
    color='blue'
)

plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='y = sin(x)')
plt.title('Multiple Annotations')

points = [(np.pi / 2, 1), (3 * np.pi / 2, -1)]
labels = ['Max', 'Min']

for (px, py), label in zip(points, labels):
    plt.annotate(
        label,
        xy=(px, py),
        xytext=(px + 0.5, py + 0.5),
        arrowprops=dict(facecolor='green', shrink=0.05),
        fontsize=10,
        color='green'
    )

plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.cos(x)

plt.plot(x, y, label='y = cos(x)')
plt.title('Cosine Curve', fontsize=16, fontweight='bold', color='purple')

plt.text(5, 0.5, 'y = cos(x)', fontsize=14, color='orange', style='italic')
plt.text(5, -0.5, 'Min = -1, Max = 1', fontsize=12, color='red')

plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='y = sin(x)')
plt.title('Highlighting a Region')

plt.axvspan(4, 6, color='yellow', alpha=0.3)
plt.text(5, 0.5, 'Highlighted Region', fontsize=12, color='blue', ha='center')

plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='y = sin(x)')
plt.title('Rotated Text Example')

plt.text(7, 0.5, 'Rotated Text', fontsize=12, color='green', rotation=45)

plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='y = sin(x)')
plt.title('Annotation with Bounding Box')

plt.text(5, 0.8, 'Important Point', fontsize=12, color='white',
         bbox=dict(facecolor='red', alpha=0.5, edgecolor='black'))

plt.legend()
plt.show()
```
