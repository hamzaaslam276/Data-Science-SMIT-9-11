```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd 
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.barplot(x="bill_length_mm",y="flipper_length_mm",data=var)

plt.title("Bar Plot of Bill_length_mm by Flipper_length_mm")


plt.show()
```

```python
sns.barplot(x="bill_length_mm",y="flipper_length_mm",data=var, orient="h" )

plt.title("Bar Plot of Bill_length_mm by Flipper_length_mm")

plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",saturation=0.1)


plt.title("Bar Plot of Island by Bill_length_mm and Sex")

plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",dodge=False)


plt.title("Bar Plot of Island by Bill_length_mm and Sex")

plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",saturation=100,alpha=0.5)

plt.title("Bar Plot of Island by Bill_length_mm and Sex")


plt.show()
```

```python
sns.set(style="darkgrid")

sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",saturation=100)

plt.title("Bar Plot of Island by Bill_length_mm and Sex")


plt.show()
```

```python
data = sns.load_dataset("titanic")

sns.barplot(data=data, x="class", y="fare")
plt.title("Basic Bar Plot")
plt.show()
```

```python
data
```

```python
sns.barplot(data=data, x="fare", y="class")
plt.title("Horizontal Bar Plot")
plt.show()
```

```python
sns.barplot(data=data, x="embark_town", y="fare", estimator=sum)
plt.title("Bar Plot with Summation Aggregation")
plt.show()
```

```python
sns.barplot(data=data, x="class", y="fare", errorbar="sd",err_kws={"color": "g","linewidth": 20})
plt.title("Bar Plot with Standard Deviation Error Bars")
plt.show()
```

```python
sns.barplot(data=data, x="class", y="fare", errorbar="sd",err_kws={"color": "g","linewidth": 10},capsize=0.5)
plt.title("Bar Plot with Standard Deviation Error Bars")
plt.show()
```

```python
sns.countplot(data=data, x="class")
plt.title("Count Plot")
plt.show()
```

```python
sns.countplot(data=data, x="class", hue="sex")
plt.title("Grouped Count Plot")
plt.show()
```

```python
sns.barplot(data=data, x="class", y="fare", width=0.5)
plt.title("Bar Plot with Custom Width")
plt.show()
```

```python
sns.barplot(data=data, x="class", y="fare", log=True)
plt.title("Bar Plot with Log Scale")
plt.show()
```
