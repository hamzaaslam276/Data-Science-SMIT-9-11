```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd 
import numpy as np
```

```python
data = sns.load_dataset("tips")
data
```

```python
sns.boxplot(data=data,x="day",y="total_bill")

plt.title("Basic Box Plot")

plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=data)

plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', hue='sex', data=data)
plt.title('Box Plot of Total Bill by Day and Sex')
plt.show()
```

```python
sns.boxplot(x='total_bill', y='day', data=data)
plt.title('Horizontal Box Plot of Total Bill by Day')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=data,hue="sex", palette='pastel')
plt.title('Box Plot with Custom Colors')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=data,hue="sex", palette='Set2')
sns.swarmplot(x='day', y='total_bill', data=data, color='.25')
plt.title('Box Plot with Swarm Plot Overlay')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=data, width=0.5)
plt.title('Box Plot with Custom Width')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=data, showfliers=True)
plt.title('Box Plot Showing Outliers')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=data, showfliers=False)
plt.title('Box Plot Without Outliers')
plt.show()
```

```python
iris = sns.load_dataset("iris")

sns.boxplot(data=iris, palette='muted')
plt.title('Box Plot for Iris Dataset')
plt.show()
```
