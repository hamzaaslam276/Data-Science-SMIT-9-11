```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd 
import numpy as np
```

```python
data = sns.load_dataset("tips")
data
```

```python
sns.barplot(x="sex",y="size",data=data)

plt.title("Basic Bar Plot")

plt.show()
```

```python
sns.countplot(x="sex",data=data,hue="smoker")

plt.title("Basic Count Plot with hue")

plt.show()
```

```python
sns.countplot(y="sex",data=data,hue="smoker")

plt.title("Basic Horizontal Count Plot with hue")

plt.show()
```

```python
sns.countplot(x="sex",data=data,hue="smoker", palette="hot" )

plt.title("Count Plot of Sex with hue and Palette ")

plt.show()
```

```python
data = sns.load_dataset("titanic")

sns.countplot(data=data, x="embark_town", hue="embark_town", palette="coolwarm", edgecolor="black", legend=False)
plt.title("Count Plot of Embarkation Town")
plt.show()
```
