```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd 
import numpy as np
```

```python
data = sns.load_dataset("tips")
data
```

```python
sns.set_style("dark")
x = sns.FacetGrid(data,col="sex",hue="day")

x.map(plt.scatter,"total_bill","tip").add_legend()

plt.show()
```
