```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd 
import numpy as np
```

```python
data = sns.load_dataset("tips")
data
```

```python
sns.catplot(x="total_bill", y="tip", data=data)

plt.show()
```

```python
sns.catplot(x="total_bill", y="tip", data=data,hue="sex",palette="hot",height=10)

plt.show()
```

```python
sns.catplot(x="total_bill", y="tip", data=data,hue="sex",palette="hot",kind="bar")

plt.show()
```

```python
sns.catplot(x="total_bill", y="tip", data=data,hue="sex",palette="hot",kind="point")

plt.show()
```

```python
sns.catplot(x="total_bill", y="tip", data=data,hue="sex",palette="hot",kind="box")

plt.show()
```

```python
sns.catplot(x="total_bill", y="tip", data=data,hue="sex",palette="hot",kind="boxen")

plt.show()
```
