```python
data2 = np.linspace(1,10,10).reshape(2,5)
data2
```

```python
sns.heatmap(data2,cmap="PuOr",annot=True)
plt.show()
```

```python
ar = np.array([['a0','a1','a2','a3','a4'],['b0','b1','b2','b3','b4']])
ar
```

```python
sns.heatmap(data2,cmap="PuOr",annot=ar,fmt="s")
plt.show()
```

```python
sns.heatmap(data2,cmap="PuOr",annot=ar,fmt="s",annot_kws=({"fontsize": 20 ,"color" : "r" }))
plt.show()
```

```python
sns.heatmap(data2,cmap="PuOr",annot=ar,fmt="s",annot_kws=({"fontsize": 20 ,"color" : "r" }),linewidths=5)
plt.show()
```

```python
sns.heatmap(data2,cmap="PuOr",annot=ar,fmt="s",annot_kws=({"fontsize": 20 ,"color" : "r" }),linewidths=5,linecolor="g")
plt.show()
```

```python
sns.heatmap(data2,cmap="PuOr",annot=ar,fmt="s",annot_kws=({"fontsize": 20 ,"color" : "r" }),linewidths=5,linecolor="g",cbar=False)
plt.show()
```

```python
v = sns.heatmap(data2,cmap="PuOr",annot=ar,fmt="s",annot_kws=({"fontsize": 20 ,"color" : "r" }),xticklabels=False,yticklabels=False)

v.set(xlabel="PYthon",ylabel="codewithHamza")

plt.show()
```

```python
v = sns.heatmap(data2,cmap="PuOr",annot=ar,fmt="s",annot_kws=({"fontsize": 20 ,"color" : "r" }),xticklabels=False,yticklabels=False)

v.set(xlabel="Python",ylabel="codewithHamza")
sns.set(font_scale=5)
plt.show()
```

```python
data = np.random.rand(5, 5) 

annotations = np.array([["a", "b", "c", "d", "e"],
                        ["f", "g", "h", "i", "j"],
                        ["k", "l", "m", "n", "o"],
                        ["p", "q", "r", "s", "t"],
                        ["u", "v", "w", "x", "y"]])

sns.heatmap(data, cmap="PuOr", annot=annotations, fmt="s", linewidths=0.5)

sns.set(font_scale=2)
plt.title("Heatmap with Custom Annotations")
plt.show()
```

```python
data = np.random.rand(10, 10)
correlation_matrix = np.corrcoef(data)

# sns.heatmap(data, annot=True, cmap="coolwarm", linewidths=0.5)

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)

plt.title("Heatmap of Random Correlation Matrix")
plt.show()
```

```python
data = sns.load_dataset("flights")

pivot_table = data.pivot(index="month", columns="year", values="passengers")

sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu", linewidths=0.5)

plt.title("Heatmap of Flights Data")
plt.show()
```
