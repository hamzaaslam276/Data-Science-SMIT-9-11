```python
data = sns.load_dataset('fmri')

sns.lineplot(data=data, x="timepoint", y="signal", hue="event", style="event")
plt.title("Basic Line Plot")
plt.show()
```

```python
data = sns.load_dataset('fmri')

sns.lineplot(data=data, x="timepoint", y="signal", hue="event", ci="sd")
plt.title("Line Plot with Confidence Intervals")
plt.show()
```

```python
sns.lineplot(data=data, x="timepoint", y="signal", hue="event", ci=None)
plt.title("Line Plot Without Confidence Intervals")
plt.show()
```

```python
sns.lineplot(
    data=data,
    x="timepoint",
    y="signal",
    hue="event",
    style="region",
    markers=True,
    dashes=False
)
plt.title("Customized Line Plot")
plt.show()
```

```python
df = pd.DataFrame({
    "time": list(range(1, 11)) * 2,
    "value": [3, 4, 5, 6, 7, 8, 9, 8, 7, 6] + [2, 3, 4, 5, 6, 7, 8, 7, 6, 5],
    "category": ["A"] * 10 + ["B"] * 10
})

sns.lineplot(data=df, x="time", y="value", hue="category")
plt.title("Line Plot with Aggregated Data")
plt.show()
```
