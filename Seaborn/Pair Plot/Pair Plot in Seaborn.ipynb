```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd 
import numpy as np
```

```python
data = sns.load_dataset("tips")
data
```

```python
sns.pairplot(data)

plt.title("Basic Pair Plot")

plt.show()
```

```python
sns.pairplot(data,hue="sex")

plt.show()
```

```python
sns.pairplot(data,vars=["total_bill","tip"],hue="sex",markers=["o", "s"])

plt.show()
```

```python
sns.pairplot(data,vars=["total_bill","tip"],hue="sex",kind="kde") # hist , kde , reg , scatter

plt.show()
```

```python
sns.pairplot(data,hue="sex",hue_order=["Female","Male"],kind="kde",diag_kind="hist") # hist , kde , reg , scatter

plt.show()
```
