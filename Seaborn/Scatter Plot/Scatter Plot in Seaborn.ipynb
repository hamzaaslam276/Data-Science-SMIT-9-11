```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd 
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.scatterplot(x="bill_length_mm",y="flipper_length_mm",data=var)

plt.title("Scatter Plot of Bill_length_mm by Flipper_length_mm")

plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="flipper_length_mm",data=var,hue="sex",size="sex",sizes=(100,50))


plt.title("Scatter Plot of Bill_length_mm by Flipper_length_mm and Sex")

plt.show()
```

```python
sns.scatterplot(data=var, x="bill_length_mm", y="bill_depth_mm", hue="species", style="island")
plt.title("Scatter Plot with Hue and Style")

plt.show()
```

```python
sns.scatterplot(data=var, x="bill_length_mm", y="bill_depth_mm", hue="species", size="body_mass_g")
plt.title("Scatter Plot with Size")
# plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
sns.lmplot(data=var, x="bill_length_mm", y="bill_depth_mm", hue="species", markers=["o", "s", "D"],palette="hot")
plt.title("Scatter Plot with Regression Line")
plt.show()
```
