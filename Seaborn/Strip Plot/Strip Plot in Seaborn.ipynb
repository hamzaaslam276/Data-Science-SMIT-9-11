```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd 
import numpy as np
```

```python
data = sns.load_dataset("tips")
data
```

```python
sns.stripplot(data=data,x="day",y="total_bill")

plt.title("Basic of Strip Plot")

plt.show()
```

```python
sns.stripplot(data=data,x="day",y="total_bill",hue="sex",palette="hot")

plt.title("Strip Plot with hue")

plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=data, jitter=True)
plt.title('Strip Plot with Jitter')
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=data, size=8, marker='o', hue='sex', palette='Set2', dodge=True)
plt.title('Strip Plot with Hue and Dodge')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=data, hue='day', palette='pastel', legend=False)
sns.stripplot(x='day', y='total_bill', data=data, jitter=True, color='black', alpha=0.5)
plt.title('Strip Plot Over Box Plot')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=data, hue='day', palette='muted', legend=False)
sns.stripplot(x='day', y='total_bill', data=data, jitter=True, color='black', alpha=0.7)
plt.title('Strip Plot Over Violin Plot')
plt.show()
```

```python
sns.stripplot(x='total_bill', y='day', data=data, jitter=True)
plt.title('Horizontal Strip Plot')
plt.show()
```
