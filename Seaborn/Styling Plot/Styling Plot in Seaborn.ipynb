```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd 
import numpy as np
```

```python
data = sns.load_dataset("tips")
data
```

```python
sns.set_style("dark")
sns.barplot(y="total_bill", x="day", data=data)

plt.show()
```

```python
sns.set_style("darkgrid")
sns.barplot(y="total_bill", x="day", data=data)

plt.show()
```

```python
sns.set_style("whitegrid")
sns.barplot(y="total_bill", x="day", data=data)

plt.show()
```

```python
sns.set_style("ticks")
sns.barplot(y="total_bill", x="day", data=data)

plt.show()
```

```python
sns.set_style("darkgrid")
sns.barplot(y="total_bill", x="day", data=data)

sns.despine()
plt.show()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12,3))

sns.barplot(y="total_bill", x="day", data=data)

sns.despine()
plt.show()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(3,5))

sns.barplot(y="total_bill", x="day", data=data)

sns.despine()
plt.show()
```

```python
sns.set_style("whitegrid")
sns.set_context("poster",font_scale=1)

sns.barplot(y="total_bill", x="day", data=data,hue="sex",palette="hot")

plt.show()
```

```python
sns.set_style("whitegrid")
sns.set_context("paper",font_scale=1)

sns.barplot(y="total_bill", x="day", data=data,hue="sex",palette="hot")

plt.show()
```
