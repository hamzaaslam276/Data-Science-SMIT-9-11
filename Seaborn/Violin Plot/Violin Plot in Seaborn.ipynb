```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd 
import numpy as np
```

```python
data = sns.load_dataset("tips")
data
```

```python
sns.violinplot(x="day",y="total_bill",data =data)

plt.title("Violin Plot of Day by Total Bill")

plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data =data,hue="time")

plt.title("Violin Plot of Day by Total Bill and Time")

plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data =data,hue="time",palette="Dark2")

plt.title("Violin Plot of Day by Total Bill and Time")

plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data =data,hue="time",linewidth=3,linecolor="g")

plt.title("Violin Plot of Day by Total Bill and Time")

plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data =data,order=["Dinner","Lunch"])

plt.title("Violin Plot of Time by Total Bill")

plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data =data,hue="sex",split=True)

plt.title("Violin Plot of Day by Total Bill and Sex")

plt.show()
```

```python
sns.violinplot(y="day",x="total_bill",data =data,hue="sex",split=True)

plt.title("Violin Plot of Day by Total Bill and Sex")

plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data =data,hue="sex",split=True,inner="quart")

plt.title("Violin Plot of Day by Total Bill and Sex")

plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data =data,order=["Dinner","Lunch"],inner="stick") # "point", "quartile",  "stick"

plt.title("Violin Plot of Time by Total Bill")

plt.show()
```

```python
sns.violinplot(x=data["total_bill"])

plt.title("Violin Plot of Total Bill")

plt.show()
```

```python
sns.violinplot(y=data["total_bill"])

plt.title("Violin Plot of Total Bill")

plt.show()
```
